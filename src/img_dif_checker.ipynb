```python
import cv2
import numpy as np

img1 = cv2.imread("spider.png")
img2 = cv2.imread("spider_img_snake_class.png")
diff = cv2.absdiff(img1, img2)
mask = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)




th = 1
imask =  mask>th

canvas = np.zeros_like(img2, np.uint8)
canvas[imask] = img2[imask]


cv2.imwrite("result.png", mask)

```

```python


img1 = cv2.imread("spider.png")
img2 = cv2.imread("spider_img_snake_class.png")
difference = cv2.subtract(img2, img1)
mask = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)


Conv_hsv_Gray = cv2.cvtColor(difference, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(Conv_hsv_Gray, 0, 255,cv2.THRESH_BINARY_INV |cv2.THRESH_OTSU)
difference[mask != 255] = [0, 0, 255]

# add the red mask to the images to make the differences obvious
img1[mask != 255] = [0, 0, 255]
img2[mask != 255] = [0, 0, 255]

# store images
cv2.imwrite('diffOverImage1.png', img1)
cv2.imwrite('diffOverImage2.png', img2)
cv2.imwrite('diff.png', difference)
```
